# Sistema de detecção de faces humanas

Nosso sistema deve ser capaz de detectar a região que representa a face, dando suporte para o sistema de classificação reconhecer a pessoa em questão.



Para o uso da Deep Learning, é necessário criar um banco de dados com vários exemplos, os quais será repassado pra IA pra que ela possa reconhecer padrões e aprender com as imagens para posteriormente conseguir identificar.

No nosso exemplo de detecção é necessário ter um banco de dados com as imagens, e em cada imagem deve-se identificar o que for necessário para o treinamento, no caso de detecção facial, temos que demarcar as posições dos rostos em cada imagem.

As demarcações podem ser feitas pelo site Labelme : http://labelme.csail.mit.edu/Release3.0/

```python
import imutils # --> biblioteca pra redimensionamento e rotação de imagens
import numpy as np # --> trabalhar com a parte matematica das imagens, vetores e matrizes
import cv2 # --> ajuda no processo de detecção da face
from google.colab.patches import cv2_imshow # --> para ver imagens na tela
from IPython.display import display, Javascript # -- > para mostrar imagens na tela
from google.colab.output import eval_js # --> para ler a web cam
from base64 import b64decode # -->converter imagens em binário
```

Código pronto para definir uma função para utilizar a WebCam para tirar uma foto


*Start* webcam

```python
def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);

      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});

      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();

      // Resize the output to fit the video element.
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);

      // Wait for Capture to be clicked.
      await new Promise((resolve) => capture.onclick = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return filename
```

Clique em 'Capturar' para fazer a foto usando sua webcam.



```python
image_file = take_photo() 

# Ao executar essa linha de codigo, abre a webCam e um botão de capturar, para
# que possamos tirar a foto, e automaticamente salva a foto no notebook.
```

Leia, redimensione e exiba a imagem.

Para que o código fique padronizado, já que dependendo da webCam a resolução e dimensionamento da foto podem variar bastante

```python
image = cv2.imread(image_file)   # Ler a Imagem que capturamos

# redimensiona para ter uma largura máxima de 400 pixels
image = imutils.resize(image, width=400)   
(h, w) = image.shape[:2]
print(w,h)

cv2_imshow(image)
```

Agora é necessário criar ou importar um modelo de Machine Learning para o Sistema.

Como criar é um processo muito demorado, inicialmente será utilizado um modelo pronto, que ja foi treinada

O detector de face em Deep Learning do OpenCV é baseado na estrutura Single Shot Detector (SSD) com uma rede base ResNet. A rede é definida e treinada usando o [Caffe Deep Learning framework](https://caffe.berkeleyvision.org/)

Baixe o modelo de detecção de rosto pré-treinado, composto por dois arquivos:

- A definição de rede (deploy.prototxt)
- Os pesos aprendidos (res10_300x300_ssd_iter_140000.caffemodel)

```python
!wget -N https://raw.githubusercontent.com/opencv/opencv/master/samples/dnn/face_detector/deploy.prototxt
!wget -N https://raw.githubusercontent.com/opencv/opencv_3rdparty/dnn_samples_face_detector_20170830/res10_300x300_ssd_iter_140000.caffemodel
```

Carregar o modelo de rede de detecção facial pré-treinado do disco




```python
print("[INFO] loading model...")  
prototxt = 'deploy.prototxt'    # A definição de rede
model = 'res10_300x300_ssd_iter_140000.caffemodel'  # pesos aprendidos
net = cv2.dnn.readNetFromCaffe(prototxt, model)
```

Use a função dnn.blobFromImage para construir um blob de entrada redimensionando a imagem para 300x300 pixels fixos e normalizando-a


A função para posicionar os demarcadores de face na sua foto precisam de posicionamento X,Y, e para padroniza-las tem-se que redimensionar, padronizar e informar os valores para o sistema. Para isso utiliza-se as linhas de código a Seguir.

```python
# redimensiona para ter uma largura máxima de 400 pixels
image = imutils.resize(image, width=400)
blob = cv2.dnn.blobFromImage(cv2.resize(image, (300, 300)), 1.0, (300, 300), (104.0, 177.0, 123.0))
```

```python
print("[INFO] computing object detections...")
net.setInput(blob)
detections = net.forward()
```

Para o caso de haver mais que apenas uma detecção, é necessário um loop para as detecções e também para que seja desenhado as caixas ao redor dos rostos detectados

```python
for i in range(0, detections.shape[2]):

	# extrair a probabilidade associada à previsão
	confidence = detections[0, 0, i, 2]

	# filtra detecções fracas garantindo que a "confiança" seja
	# maior que o limite mínimo de confiança
	if confidence > 0.5: #Nossa detecção deve ter no mínimo 50% de certeza
		# calcula as coordenadas (x, y) da caixa delimitadora do objeto
		box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
		(startX, startY, endX, endY) = box.astype("int")
		# desenha a caixa delimitadora da face junto com a probabilidade associada
		text = "{:.2f}%".format(confidence * 100)
		y = startY - 10 if startY - 10 > 10 else startY + 10
		cv2.rectangle(image, (startX, startY), (endX, endY), (0, 0, 255), 2)
		cv2.putText(image, text, (startX, y),
			cv2.FONT_HERSHEY_SIMPLEX, 0.45, (0, 0, 255), 2)
```
